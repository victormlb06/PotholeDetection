```python
import cv2
```

```python
VIDEO_PATH = '/content/drive/My Drive/DeepLearning-UFG/Projeto/Potholev1/Video/Reportagem.mp4'
VIDEO_OUT = '/content/drive/My Drive/DeepLearning-UFG/Projeto/Potholev1/Video/'
```

```python
cap = cv2.VideoCapture(VIDEO_PATH)
```

```python
ret, frame = cap.read()
```

```python
H, W, _ = frame.shape
```

```python
out = cv2.VideoWriter(VIDEO_OUT, cv2.VideoWriter_fourcc(*'MP4V'),int(cap.get(cv2.CAP_PROP_FPS)),(W,H))
```

```python
model = YOLO("/content/drive/MyDrive/DeepLearning-UFG/Projeto/Potholev2Modelos/Final_PD3/weights/best.pt")
```

```python
threshold = 0.5
```

```python
while ret:

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
