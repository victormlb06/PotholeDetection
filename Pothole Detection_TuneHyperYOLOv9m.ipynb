```python
from ultralytics import YOLO

# Initialize the YOLO model
model = YOLO('yolov9m.yaml')

# Tune hyperparameters on COCO8 for 30 epochs
model.tune(data=os.path.join(ROOT_DIR,"dataColab.yaml"),
                  epochs=30, iterations=30,
                  optimizer='AdamW',
                  plots=True,
                  save=True,
                  val=True,
                  batch = -1,
                  device = 'cuda'
          )
```
